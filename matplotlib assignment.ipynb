## What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

### Ans: Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of functions and classes that enable users to generate various types of plots, charts, and graphs.
Matplotlib is commonly used for data visualization in fields such as data science, machine learning, statistics, and scientific research. It offers a flexible and customizable framework for creating publication-quality visualizations, allowing users to convey complex data patterns and insights effectively.
The Pyplot module is a sublibrary of Matplotlib that provides a simplified interface for creating basic plots and figures. It is widely used for quick data visualization tasks and prototyping. Here are five types of plots that can be plotted using the Pyplot module:
1. Line Plot: Line plots are used to display the relationship between two continuous variables. They are suitable for showing trends and patterns over time or any other continuous axis.
2. Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. They represent individual data points as dots and can help identify patterns, clusters, or correlations between variables.
3. Bar Plot: Bar plots (also known as bar charts) are used to compare categorical data or to display the frequency or distribution of a categorical variable. They use rectangular bars to represent the values of each category.
4. Histogram: Histograms are used to represent the distribution of a continuous variable. They group data into bins and display the frequency or count of values falling within each bin.
5. Pie Chart: Pie charts are used to represent the proportions or percentages of different categories in a dataset. They divide a circle into sectors, with each sector representing a category and its size proportional to the corresponding value.

## What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

### Ans: A scatter plot is a type of plot used to display the relationship between two continuous variables. It represents individual data points as dots on a two-dimensional graph, with one variable plotted along the x-axis and the other variable plotted along the y-axis. Scatter plots are useful for visualizing patterns, clusters, or correlations between variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)  
plt.title('Scatter Plot') 
plt.xlabel('X-axis') 
plt.ylabel('Y-axis')  
plt.show()  
```

# Question.3

## Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

### Ans: The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to create multiple plots or charts in a single figure window, arranged in rows and columns. This function is useful when you want to compare or visualize multiple plots side by side.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Line 1')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Line 2')
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Line 3')
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Line 4')
plt.tight_layout()
plt.show()
```

# Question.4

## What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

### Ans: A bar plot, also known as a bar chart, is a type of plot that uses rectangular bars to represent the values of a categorical variable. The length or height of each bar corresponds to the value of the variable being plotted. Bar plots are used to compare and visualize the values or frequencies of different categories or groups.

Bar plots are particularly useful when you want to:

Compare the magnitudes of different categories or groups.
Visualize the distribution or frequency of categorical data.
Highlight trends or patterns across categories.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

## What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

### Ans: A box plot, also known as a box-and-whisker plot, is a type of plot used to visualize the distribution of a dataset through quartiles. It provides a summary of the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values of a dataset. The plot consists of a rectangular box (from Q1 to Q3), a line at the median, and "whiskers" extending from the box to represent the minimum and maximum values.

Box plots are used to:

Identify the central tendency and spread of a dataset.
Detect outliers or extreme values in the data.
Compare the distributions of multiple datasets.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
