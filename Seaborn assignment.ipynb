## Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### Ans: Seaborn is a powerful Python library for statistical data visualization. It provides various types of plots to explore and analyze data. Here are five common plots that can be created using Seaborn and their uses:
1. **Scatter Plot**: A scatter plot displays the relationship between two continuous variables. It is useful for identifying patterns, trends, and correlations between variables.
2. **Histogram**: A histogram visualizes the distribution of a single continuous variable. It helps understand the frequency or count of values within specific bins or ranges.
3. **Bar Plot**: A bar plot compares categorical variables and their associated values. It is suitable for displaying the distribution or summary statistics of categorical data.
4. **Box Plot**: A box plot shows the distribution of a numerical variable across different categories. It provides information about the median, quartiles, outliers, and skewness of the data.
5. **Heatmap**: A heatmap displays a matrix of values as a grid of colored cells. It is useful for visualizing correlations, relationships, or patterns in numerical data.
Uses of each plot:
- Scatter Plot: Useful for visualizing relationships and patterns between continuous variables, identifying clusters or outliers, and assessing correlation strength.
- Histogram: Useful for understanding the distribution, skewness, and outliers of a single variable.
- Bar Plot: Useful for comparing categorical variables and their associated values, visualizing frequency or proportions, and identifying dominant categories.
- Box Plot: Useful for understanding the distribution, variability, and skewness of a numerical variable across different groups or categories.
- Heatmap: Useful for visualizing relationships, patterns, or correlations between variables in a matrix-like dataset.

## 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
### Ans:
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("fmri")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="timepoint", y="signal", hue="event", style="region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Signal over Time for Different Events and Regions")
plt.show()
```

# Question.3

## Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset from Seaborn
df = sns.load_dataset("titanic")

# Create the box plots
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

## Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the due parameter for the 'cut' column of the diamonds dataset.

```python
##Answer:
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="price", hue="cut", multiple="stack")
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")
plt.show()
```
